# Project Context  
The dataset I have chosen contains information on more than 20,000 chess games, collected via the Lichess.org API, a highly popular online platform. Each record in the dataset represents a chess game and includes important details such as:

- The Elo ratings of the players (white and black), which measure their skill level at the start of each game.  
- The outcome of the game, indicating whether the white, black, or neither (draw) side won.  
- The number of moves played during the game.  
- The openings used by the players, as well as information on the game's progression.  
- The player IDs, allowing identification of games played by specific individuals.  

This project revolves around analyzing this data to extract relevant insights, such as victory patterns, result distribution based on player groups, and the impact of certain parameters like openings or the number of moves.

These analyses will help identify key factors that influence the outcomes of chess games, allowing for the discovery of significant patterns that can provide players with a strategic advantage.

```python
mean_white_rating = df_clean['white_rating'].mean()
std_white_rating = df_clean['white_rating'].std()

mean_black_rating = df_clean['black_rating'].mean()
std_black_rating = df_clean['black_rating'].std()

print(f"Mean Elo of Whites: {mean_white_rating:.2f} (Standard Deviation: {std_white_rating:.2f})")
print(f"Mean Elo of Blacks: {mean_black_rating:.2f} (Standard Deviation: {std_black_rating:.2f})")
```

## Analysis of Turns Required to Win  
For this analysis, we sought to understand how many turns are necessary for a player to win a chess game, separating the wins of white and black players. We began by defining the essential elements to calculate these statistics.

- <span style="color:red">white_wins</span> and <span style="color:red">black_wins</span> were defined by filtering the cleaned DataFrame <span style="color:red">df_clean</span> to isolate the games won by white and black, respectively:  
  This allows us to analyze the performances of both colors separately.

- Next, for each group (white and black), we calculated three key statistics on the <span style="color:red">turns</span> column (number of turns):

    - The <span style="color:red">mean</span>: the average number of turns to finish a game.
    - The <span style="color:red">median</span>: the central number of turns, indicating where the majority of games lie.
    - The <span style="color:red">standard deviation</span> (std): which measures the dispersion of values around the mean.

These definitions allowed us to observe that white wins on average in <span style="color:red">57.76 turns</span>, while black requires <span style="color:red">60.73 turns</span>. Although this data is interesting, the gap between the two is too small to draw a significant conclusion or identify a useful pattern in our search for factors that increase the chances of victory.


```python
white_wins = df_clean[df_clean['winner'] == 'white']

black_wins = df_clean[df_clean['winner'] == 'black']

white_turns_mean = white_wins['turns'].mean()
white_turns_median = white_wins['turns'].median()
white_turns_std = white_wins['turns'].std()

black_turns_mean = black_wins['turns'].mean()
black_turns_median = black_wins['turns'].median()
black_turns_std = black_wins['turns'].std()

print(f"Mean turns for Whites: {white_turns_mean:.2f} (Median: {white_turns_median}, Standard Deviation: {white_turns_std:.2f})")
print(f"Mean turns for Blacks: {black_turns_mean:.2f} (Median: {black_turns_median}, Standard Deviation: {black_turns_std:.2f})")
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(white_wins['turns'], bins=30, color='blue', label='Whites', kde=True)
sns.histplot(black_wins['turns'], bins=30, color='orange', label='Blacks', kde=True)
plt.title('Comparison of Turns Required for Wins in Chess')
plt.xlabel('Number of Turns')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

## Analysis of Winning Players' Elo  
In this section, we aimed to analyze the Elo of players who won games. To do this, we first extracted the necessary information from the DataFrame:

I used the <span style="color:red">apply()</span> method with a lambda function to create a new column <span style="color:red">winning_rating</span>. This column contains the Elo of the winning player:

- If the winning player is white, I retrieve <span style="color:red">white_rating</span>.
- If the winning player is black, I retrieve <span style="color:red">black_rating</span>.

Once the data was prepared, I created a <span style="color:red">histogram</span> to visualize the distribution of the winning players' Elo:

The chart shows the distribution of the Elo of players who have won a game. This allows us to see which types of players (in terms of level) win most often.

After conducting this analysis, I concluded that it was necessary to examine the number of players by Elo ranking to determine if these figures could reveal significant information. This is what I did next.

*Brief explanation of the lambda formula: I used it to automatically select the Elo of the winning player, whether they are white or black, without having to manually separate the two groups.*


```python
winning_ratings = df_clean[['winner', 'white_rating', 'black_rating']].copy()  # Create a copy

winning_ratings['winning_rating'] = winning_ratings.apply(
    lambda row: row['white_rating'] if row['winner'] == 'white' else row['black_rating'], axis=1
)

plt.figure(figsize=(10, 6))
sns.histplot(winning_ratings['winning_rating'], bins=30, color='purple', kde=True)
plt.title('Elo of Winning Players in Chess Games')
plt.xlabel('Elo')
plt.ylabel('Frequency')
plt.show()
```

## Analysis of Players' Elo to Rank Them by Skill Level  
The function <span style="color:red">categorize_rating</span> is defined to categorize players' Elo. It assigns the category <span style="color:red">Beginner</span> to players with an Elo below 1200, <span style="color:red">Intermediate</span> for those with an Elo between 1200 and 1800, and <span style="color:red">Advanced</span> for those with an Elo above 1800.

Next, I apply this function to the <span style="color:red">white_rating</span> column of the <span style="color:red">df_clean</span> DataFrame to create a new column called <span style="color:red">rating_category</span>. This allows me to classify each player based on their Elo.

To visualize the distribution of players in each category, I count the number of players per category using <span style="color:red">value_counts()</span>, then transform the result into a DataFrame.

The analysis of Elo reveals a majority of players categorized as Intermediate, with a much smaller number of Beginners and Advanced players. This suggests that there is a strong concentration of average players in the player pool, which can influence the dynamics of the games, as Intermediate players are very heterogeneous in their skill levels.


```python
def categorize_rating(rating):
    if rating < 1200:
        return 'Beginner'
    elif 1200 <= rating < 1800:
        return 'Intermediate'
    else:
        return 'Advanced'

df_clean.loc[:, 'rating_category'] = df_clean['white_rating'].apply(categorize_rating)

category_counts = df_clean['rating_category'].value_counts().reset_index()
category_counts.columns = ['rating_category', 'number_of_players']

plt.figure(figsize=(10, 6))
sns.barplot(x='rating_category', y='number_of_players', hue='rating_category', data=category_counts, palette='viridis', legend=False)

plt.title('Number of Players by Elo')
plt.xlabel('Rank')
plt.ylabel('Number of Players')
plt.xticks(rotation=30) 
plt.show()
```

## Analysis of Wins by Elo Group
In this analysis, we compared the number of wins for white and black players based on their Elo (<span style="color:red">beginner</span>, <span style="color:red">intermediate</span>, <span style="color:red">advanced</span>). The goal was to examine if certain groups had more success than others and to determine if there are notable trends in wins by skill level.

First, I extracted and counted the wins for whites and blacks by filtering the DataFrame for each color. For whites, I counted wins in the <span style="color:red">white_group</span> column, and for blacks in the <span style="color:red">black_group</span> column. These two columns contain the classifications of players (<span style="color:red">beginner</span>, <span style="color:red">intermediate</span>, <span style="color:red">advanced</span>), which I previously created using the <span style="color:red">categorize_rating</span> function. The counting was done using the <span style="color:red">value_counts()</span> method, allowing us to obtain the number of wins for each rank. Then, I created a DataFrame <span style="color:red">group_results</span> to combine the win results for whites and blacks into a table.

By visualizing the results with a bar graph, we can observe the distribution of wins by rank for both colors. This graph highlights the performance differences between player groups and allows us to assess whether a certain skill level tends to win more games.

However, while this graph provides information on the distribution of wins, it does not necessarily reveal an obvious <span style="color:red">winning pattern</span>. It remains necessary to complement this analysis with other factors to better understand the dynamics of victory.


```python
def categorize_rating(rating):
    if rating < 1200:
        return 'Beginner'
    elif 1200 <= rating < 1800:
        return 'Intermediate'
    else:
        return 'Advanced'

df_clean = df_clean.copy()
df_clean.loc[:, 'white_group'] = df_clean['white_rating'].apply(categorize_rating)
df_clean.loc[:, 'black_group'] = df_clean['black_rating'].apply(categorize_rating)

white_group_wins = df_clean[df_clean['winner'] == 'white']['white_group'].value_counts()
black_group_wins = df_clean[df_clean['winner'] == 'black']['black_group'].value_counts()

group_results = pd.DataFrame({
    'White Wins': white_group_wins,
    'Black Wins': black_group_wins
})  
print(group_results)

group_results.plot(kind='bar', figsize=(10, 6))
plt.title('Wins by Elo Group')
plt.xlabel('Elo Group')
plt.ylabel('Number of Wins')
plt.xticks(rotation=45)
plt.legend(title='Winning Color')
plt.show()
```
