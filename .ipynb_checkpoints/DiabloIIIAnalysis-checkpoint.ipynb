# Diablo III Data Analysis Project

Welcome to the **Diablo III Data Analysis** project! In this notebook, we will explore and analyze various aspects of the Diablo III API. From leaderboard analysis to hero statistics and item breakdowns, we aim to uncover insights about top-performing players and classes based on Rift Times across different seasons.

This notebook will guide you through:

- Fetching and analyzing season data
- Displaying leaderboards for different classes
- Exploring the most used items by top players
- Visualizing class performance based on Rift Times

Let's dive into the analysis!

## 1. Import Libraries

First, we need to import the necessary libraries for data manipulation, visualization, and API interactions.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import requests
from datetime import timedelta
from IPython.display import display, Markdown, clear_output
from ipywidgets import interact, widgets, Button, Output
import json

# Set the base URL for the API
BASE_URL = "http://localhost:5000/api"  # Update this with the correct URL if different

# Set the style for plots
sns.set(style="whitegrid")
```

## 2. Fetch Seasons Data from the API

We will start by fetching the available seasons from the Diablo III API.



```python
def fetch_seasons():
    url = f"{BASE_URL}/seasons"
    response = requests.get(url)

    if response.status_code == 200:
        try:
            return response.json()  # Ensure it returns parsed JSON data
        except ValueError:
            print("Error: Unable to parse response as JSON.")
            return []
    else:
        print(f"Failed to fetch seasons data: {response.status_code}")
        return []
```

## 3. Display Seasons Data

We will display the seasons data in a table format and add interactive buttons to navigate through the data.


```python
# Function to display seasons data with centered columns
def display_seasons_data(start_index=0):
    seasons_data = fetch_seasons()
    if seasons_data:
        seasons = [season['href'].split('?')[0].split('/')[-1] for season in seasons_data['season']]
        display(Markdown("### Seasons Data"))
        display(pd.DataFrame(seasons[start_index:start_index+10], columns=['Season']).style.set_properties(**{'text-align': 'center'}))
    else:
        display(Markdown("### No seasons data available"))

# Interactive buttons for seasons
seasons_output = Output()
seasons_start_index = 0

def on_seasons_next_click(b):
    global seasons_start_index
    seasons_start_index += 10
    with seasons_output:
        clear_output(wait=True)
        display_seasons_data(seasons_start_index)

def on_seasons_prev_click(b):
    global seasons_start_index
    seasons_start_index = max(0, seasons_start_index - 10)
    with seasons_output:
        clear_output(wait=True)
        display_seasons_data(seasons_start_index)

seasons_next_button = Button(description="Next")
seasons_prev_button = Button(description="Previous")
seasons_next_button.on_click(on_seasons_next_click)
seasons_prev_button.on_click(on_seasons_prev_click)

with seasons_output:
    display_seasons_data(seasons_start_index)

display(seasons_output, seasons_prev_button, seasons_next_button)
```

## 4. Fetch and Display Leaderboard Data

We will fetch leaderboard data for all classes and display the top 150 players sorted by the lowest Rift Time.


```python
# Function to convert Rift Time from milliseconds to mm:ss format
def convert_time_to_mmss(milliseconds):
    if milliseconds is not None:
        seconds = milliseconds / 1000  # Convert milliseconds to seconds
        minutes, seconds = divmod(seconds, 60)  # Get minutes and seconds
        return f"{int(minutes):02}:{int(seconds):02}"  # Format as mm:ss
    return "N/A"

# Function to fetch leaderboard data from the API based on the season and class
def fetch_leaderboard(season_id, leaderboard_name):
    url = f"{BASE_URL}/leaderboard?season_id={season_id}&leaderboard_name={leaderboard_name}"
    response = requests.get(url)

    if response.status_code == 200:
        return response.json()
    else:
        print(f"Failed to fetch leaderboard data: {response.status_code}")
        return []

# Function to extract leaderboard info into a structured DataFrame
def extract_leaderboard_info(data, class_name):
    class_name_mapping = {
        "rift-barbarian": "Barbarian",
        "rift-wizard": "Wizard",
        "rift-dh": "Demon Hunter",
        "rift-monk": "Monk",
        "rift-necromancer": "Necromancer",
        "rift-wd": "Witch Doctor",
        "rift-crusader": "Crusader"
    }

    friendly_class_name = class_name_mapping.get(class_name, class_name)

    if 'row' in data:
        rows = data['row']
        extracted_data = []
        for row in rows:
            row_data = {entry['id']: entry for entry in row['data']}
            player_data = row['player'][0]['data']
            player_info = {entry['id']: entry for entry in player_data}
            extracted_data.append({
                "Rank": row_data.get('Rank', {}).get('number', None),
                "BattleTag": row_data.get('BattleTag', {}).get('string', None),
                "Rift Level": row_data.get('RiftLevel', {}).get('number', None),
                "Rift Time (ms)": row_data.get('RiftTime', {}).get('timestamp', None),
                "Rift Time (mm:ss)": convert_time_to_mmss(row_data.get('RiftTime', {}).get('timestamp')),
                "Class": friendly_class_name,
                "Paragon Level": player_info.get('ParagonLevel', {}).get('number', None),
                "Clan Name": player_info.get('ClanName', {}).get('string', None),
                "Hero Level": player_info.get('HeroLevel', {}).get('number', None),
                "Hero Gender": player_info.get('HeroGender', {}).get('string', None)
            })
        return pd.DataFrame(extracted_data)
    return pd.DataFrame()

# Function to fetch leaderboard data for all classes and combine them into one DataFrame
def fetch_leaderboard_data_for_all_classes(season_id):
    classes = ["rift-barbarian", "rift-wizard", "rift-dh", "rift-monk", "rift-necromancer", "rift-wd", "rift-crusader"]
    leaderboard_data = []
    
    for class_name in classes:
        data = fetch_leaderboard(season_id, class_name)
        if data:
            leaderboard_data.append(extract_leaderboard_info(data, class_name))
    
    # Combine all dataframes into one
    return pd.concat(leaderboard_data, ignore_index=True)

# Function to display the top 150 leaderboard data sorted by the lowest Rift Time
def display_top_150_leaderboard(season_id):
    leaderboard_data = fetch_leaderboard_data_for_all_classes(season_id)
    
    if not leaderboard_data.empty:
        # Sort by Rift Time in milliseconds (ascending)
        leaderboard_data_sorted = leaderboard_data.sort_values(by='Rift Time (ms)', ascending=True)
        
        # Display the top 150 entries
        display(Markdown(f"### Top 150 Leaderboard Data for Season {season_id}, Sorted by Lowest Rift Time"))
        display(leaderboard_data_sorted.head(150).style.set_properties(**{'text-align': 'center'}))
    else:
        display(Markdown("### No leaderboard data available"))

# Example usage: Display top 150 players for Season 23
display_top_150_leaderboard(32)
```

## 5. Visualize Lowest Rift Time Per Class

Let's visualize the lowest Rift Time per class to see which class performs the fastest.


```python
def plot_lowest_rift_time_per_class(season_id):
    leaderboard_data = fetch_leaderboard_data_for_all_classes(season_id)
    
    if not leaderboard_data.empty:
        # Find the lowest Rift Time per class
        lowest_rift_time_per_class = leaderboard_data.groupby('Class').agg({'Rift Time (ms)': 'min'}).reset_index()

        # Convert Rift Time from milliseconds to mm:ss for display
        lowest_rift_time_per_class['Rift Time (mm:ss)'] = lowest_rift_time_per_class['Rift Time (ms)'].apply(convert_time_to_mmss)

        # Plot the comparison using a bar chart
        plt.figure(figsize=(10, 6))
        sns.barplot(x='Class', y='Rift Time (ms)', data=lowest_rift_time_per_class, hue='Class', palette='viridis', dodge=False, legend=False)
        plt.title(f'Lowest Rift Time Per Class for Season {season_id}')
        plt.xlabel('Class')
        plt.ylabel('Rift Time (ms)')
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()

        # Display the data
        display(Markdown(f"### Lowest Rift Time per Class for Season {season_id}"))
        display(lowest_rift_time_per_class)
        
    else:
        display(Markdown("### No leaderboard data available"))

# Example usage: Plot lowest Rift Time for Season 23
plot_lowest_rift_time_per_class(32)
```

## 6. Visualize Rift Level Distribution

We can also visualize the distribution of Rift Levels achieved by players in the top 150.


```python
def plot_rift_level_distribution(season_id):
    leaderboard_data = fetch_leaderboard_data_for_all_classes(season_id)
    if not leaderboard_data.empty:
        # Filter top 150 players and make a copy
        leaderboard_data_sorted = leaderboard_data.sort_values(by='Rift Time (ms)', ascending=True)
        top_150 = leaderboard_data_sorted.head(150).copy()
        
        # Convert 'Rift Level' to string type
        top_150['Rift Level'] = top_150['Rift Level'].astype(str)
        
        # Plot the Rift Level distribution
        plt.figure(figsize=(10, 6))
        sns.countplot(
            x='Rift Level',
            data=top_150,
            hue='Rift Level',
            palette='coolwarm',
            dodge=False,
            order=sorted(top_150['Rift Level'].unique()),
            legend=False
        )
        plt.title(f'Rift Level Distribution among Top 150 Players for Season {season_id}')
        plt.xlabel('Rift Level')
        plt.ylabel('Number of Players')
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()
    else:
        display(Markdown("### No leaderboard data available"))

# Example usage: Plot Rift Level distribution for Season 23
plot_rift_level_distribution(32)
```

## 7. Visualize Class Representation in Top 150

We can visualize how many players of each class are present in the top 150.


```python
def plot_class_representation(season_id):
    leaderboard_data = fetch_leaderboard_data_for_all_classes(season_id)
    if not leaderboard_data.empty:
        # Filter top 150 players and make a copy
        leaderboard_data_sorted = leaderboard_data.sort_values(by='Rift Time (ms)', ascending=True)
        top_150 = leaderboard_data_sorted.head(150).copy()
        
        # Convert 'Class' to string type
        top_150['Class'] = top_150['Class'].astype(str)
        
        # Plot the class representation
        plt.figure(figsize=(10, 6))
        sns.countplot(
            x='Class',
            data=top_150,
            hue='Class',
            palette='Set2',
            dodge=False,
            legend=False
        )
        plt.title(f'Class Representation among Top 150 Players for Season {season_id}')
        plt.xlabel('Class')
        plt.ylabel('Number of Players')
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()
    else:
        display(Markdown("### No leaderboard data available"))


# Example usage: Plot class representation for Season 23
plot_class_representation(32)
```

## 8. Fetch and Display Player Information

We will now fetch detailed information about players, including their heroes and items.


```python
# Function to convert Rift Time from milliseconds to mm:ss format
def convert_time_to_mmss(milliseconds):
    if milliseconds is not None:
        seconds = milliseconds / 1000  # Convert milliseconds to seconds
        minutes, seconds = divmod(seconds, 60)  # Get minutes and seconds
        return f"{int(minutes):02}:{int(seconds):02}"  # Format as mm:ss
    return "N/A"

# Function to fetch leaderboard data from the API based on the season and class
def fetch_leaderboard(season_id, leaderboard_name):
    url = f"{BASE_URL}/leaderboard?season_id={season_id}&leaderboard_name={leaderboard_name}"
    response = requests.get(url)
    if response.status_code == 200:
        return response.json()
    else:
        print(f"Failed to fetch leaderboard data: {response.status_code}")
        return []

def extract_leaderboard_info(data, class_name):
    class_name_mapping = {
        "rift-barbarian": "Barbarian",
        "rift-wizard": "Wizard",
        "rift-dh": "Demon Hunter",
        "rift-monk": "Monk",
        "rift-necromancer": "Necromancer",
        "rift-wd": "Witch Doctor",
        "rift-crusader": "Crusader"
    }

    friendly_class_name = class_name_mapping.get(class_name, class_name)

    if 'row' in data:
        rows = data['row']
        extracted_data = []
        for row in rows:
            row_data = {entry['id']: entry for entry in row['data']}

            # Initialize variables
            battletag = row_data.get('BattleTag', {}).get('string', None)
            hero_id = None
            paragon_level = None
            clan_name = None
            hero_level = None
            hero_gender = None

            if 'player' in row and row['player']:
                player_data = row['player'][0]['data']
                player_info = {entry['id']: entry for entry in player_data}

                # Try to extract BattleTag from player_info if missing
                if not battletag:
                    battletag = player_info.get('HeroBattleTag', {}).get('string', None)

                hero_id = player_info.get('HeroId', {}).get('number', None)
                paragon_level = player_info.get('ParagonLevel', {}).get('number', None)
                clan_name = player_info.get('ClanName', {}).get('string', None)
                hero_level = player_info.get('HeroLevel', {}).get('number', None)
                hero_gender = player_info.get('HeroGender', {}).get('string', None)

            extracted_data.append({
                "Rank": row_data.get('Rank', {}).get('number', None),
                "BattleTag": battletag,
                "HeroId": hero_id,
                "Rift Level": row_data.get('RiftLevel', {}).get('number', None),
                "Rift Time (ms)": row_data.get('RiftTime', {}).get('timestamp', None),
                "Rift Time (mm:ss)": convert_time_to_mmss(row_data.get('RiftTime', {}).get('timestamp')),
                "Class": friendly_class_name,
                "Paragon Level": paragon_level,
                "Clan Name": clan_name,
                "Hero Level": hero_level,
                "Hero Gender": hero_gender
            })
        return pd.DataFrame(extracted_data)
    return pd.DataFrame()



# Function to fetch leaderboard data for all classes and combine them into one DataFrame
def fetch_leaderboard_data_for_all_classes(season_id):
    classes = ["rift-barbarian", "rift-wizard", "rift-dh", "rift-monk", "rift-necromancer", "rift-wd", "rift-crusader"]
    leaderboard_data = []

    for class_name in classes:
        data = fetch_leaderboard(season_id, class_name)
        if data:
            leaderboard_data.append(extract_leaderboard_info(data, class_name))

    # Combine all dataframes into one
    return pd.concat(leaderboard_data, ignore_index=True)

# Generalized fetch function with error handling
def fetch_data(url):
    response = requests.get(url)
    if response.status_code == 200:
        try:
            return response.json()
        except ValueError:
            print(f"Error parsing JSON response from {url}")
            return {}
    else:
        print(f"Failed to fetch data from {url}: {response.status_code}")
        return {}

# Function to fetch account profile
def fetch_account_profile(account):
    url = f"{BASE_URL}/account/{account}"
    return fetch_data(url)

# Function to display a structured account profile
# Function to display a structured account profile
def display_account_profile(account):
    profile_data = fetch_account_profile(account)
    if profile_data:
        display(Markdown(f"## Account Profile for {account}"))

        # Extract and display the main profile info
        profile_info = {
            "BattleTag": profile_data.get('battleTag', 'N/A'),
            "Guild Name": profile_data.get('guildName', 'N/A'),
            "Paragon Level": profile_data.get('paragonLevel', 'N/A'),
            "Paragon Level Hardcore": profile_data.get('paragonLevelHardcore', 'N/A'),
            "Paragon Level Season": profile_data.get('paragonLevelSeason', 'N/A'),
            "Paragon Level Season Hardcore": profile_data.get('paragonLevelSeasonHardcore', 'N/A'),
            "Highest Hardcore Level": profile_data.get('highestHardcoreLevel', 'N/A'),
            "Last Hero Played": profile_data.get('lastHeroPlayed', 'N/A'),
            "Last Updated": profile_data.get('lastUpdated', 'N/A'),
        }
        display(pd.DataFrame([profile_info]))

        # Display artisan levels
        artisans = ['blacksmith', 'jeweler', 'mystic']
        artisan_info = {}
        for artisan in artisans:
            artisan_data = profile_data.get(artisan, {})
            artisan_info.update({
                f"{artisan.capitalize()} Level": artisan_data.get('level', 'N/A'),
            })

        display(Markdown("### Artisan Levels"))
        display(pd.DataFrame([artisan_info]))

        # Display kills
        kills = profile_data.get('kills', {})
        kills_formatted = {f"Kills {k.capitalize()}": v for k, v in kills.items()}
        display(Markdown("### Kills"))
        display(pd.DataFrame([kills_formatted]))

        # Display time played
        time_played = profile_data.get('timePlayed', {})
        time_played_percent = {f"Time Played {k.capitalize()}": f"{v*100:.2f}%" for k, v in time_played.items()}
        display(Markdown("### Time Played"))
        display(pd.DataFrame([time_played_percent]))

        # Display heroes summary
        display(Markdown(f"### Heroes"))
        heroes = profile_data.get('heroes', [])
        if heroes:
            heroes_df = pd.DataFrame(heroes)[['id', 'name', 'class', 'level', 'paragonLevel', 'seasonal', 'hardcore', 'gender', 'last-updated']]
            display(heroes_df)
        else:
            display(Markdown("No heroes found for this account."))

        # Display seasonal profiles
        seasonal_profiles = profile_data.get('seasonalProfiles', {})
        if seasonal_profiles:
            display(Markdown("### Seasonal Profiles"))
            for season, season_data in seasonal_profiles.items():
                display(Markdown(f"#### {season.capitalize()}"))
                season_info = {
                    "Season ID": season_data.get('seasonId', 'N/A'),
                    "Paragon Level": season_data.get('paragonLevel', 'N/A'),
                    "Paragon Level Hardcore": season_data.get('paragonLevelHardcore', 'N/A'),
                    "Highest Hardcore Level": season_data.get('highestHardcoreLevel', 'N/A'),
                }

                # Process kills data
                kills_data = season_data.get('kills', {})
                kills_data_formatted = {f"Kills {k.capitalize()}": v for k, v in kills_data.items()}

                # Process time played data
                time_played_data = season_data.get('timePlayed', {})
                time_played_data_formatted = {f"Time Played {k.capitalize()}": f"{v*100:.2f}%" for k, v in time_played_data.items()}

                # Combine all data into one dictionary
                season_info.update(kills_data_formatted)
                season_info.update(time_played_data_formatted)

                # Display the season info as a DataFrame
                display(pd.DataFrame([season_info]))
        else:
            display(Markdown("No seasonal profiles found."))

    else:
        display(Markdown("### No account profile available"))


# Function to fetch hero items
def fetch_hero_items(account, hero_id):
    if hero_id == "Unknown":
        print(f"Skipping hero items fetch as hero ID is 'Unknown'.")
        return {}

    url = f"{BASE_URL}/character/{account}/hero/{hero_id}/items"
    return fetch_data(url)

# Flatten item attributes for display
def flatten_item_attributes(attributes):
    primary_attributes = attributes.get('primary', [])
    secondary_attributes = attributes.get('secondary', [])
    return {
        "Primary Attributes": ', '.join(primary_attributes),
        "Secondary Attributes": ', '.join(secondary_attributes)
    }

# Function to fetch hero data
def fetch_hero_data(account, hero_id):
    if hero_id == "Unknown":
        print(f"Skipping hero data fetch as hero ID is 'Unknown'.")
        return {}
    url = f"{BASE_URL}/character/{account}/hero/{hero_id}"
    return fetch_data(url)

def display_player_info(account, hero_id, hero_class):
    # Display Account Profile
    display_account_profile(account)

    # Fetch and display hero data
    if hero_id != "Unknown":
        hero_data = fetch_hero_data(account, hero_id)
        if hero_data:
            display(Markdown(f"## Hero Data for {account}, Hero ID: {hero_id}"))
            # Extract and display hero information
            hero_info = {
                "Name": hero_data.get('name', 'Unknown'),
                "Class": hero_data.get('class', 'Unknown'),
                "Gender": 'Male' if hero_data.get('gender', 0) == 0 else 'Female',
                "Level": hero_data.get('level', 'Unknown'),
                "Paragon Level": hero_data.get('paragonLevel', 'Unknown'),
                "Hardcore": hero_data.get('hardcore', False),
                "Seasonal": hero_data.get('seasonal', False),
                "Last Updated": hero_data.get('last-updated', 'Unknown'),
                "Elite Kills": hero_data.get('kills', {}).get('elites', 'N/A'),
            }
            display(pd.DataFrame([hero_info]))

            # Fetch and display hero items
            hero_items_data = fetch_hero_items(account, hero_id)
            if hero_items_data and isinstance(hero_items_data, dict):
                display(Markdown(f"### Hero Items for {account}, Hero ID: {hero_id}"))
                items_flat = []
                for slot, item in hero_items_data.items():
                    if not isinstance(item, dict):
                        print(f"Invalid item data in slot '{slot}': {item}")
                        continue  # Skip or handle accordingly

                    item_attributes = flatten_item_attributes(item.get('attributes', {}))
                    item_data = {
                        "Slot": slot,
                        "Name": item.get('name', 'Unknown'),
                        "Type": item.get('typeName', 'Unknown'),
                        "Item Level": item.get('itemLevel', 'Unknown'),
                        "Primary Attributes": item_attributes.get("Primary Attributes", ""),
                        "Secondary Attributes": item_attributes.get("Secondary Attributes", ""),
                        "Set": item.get('set', {}).get('name', 'None'),
                        "Legendary Effects": item.get('flavorText', 'None'),
                    }
                    items_flat.append(item_data)
                if items_flat:
                    df_items = pd.DataFrame(items_flat)
                    display(df_items)
                else:
                    display(Markdown(f"No valid items found for {account}, Hero ID: {hero_id}."))
            else:
                display(Markdown(f"### No items data available for {account}, Hero ID: {hero_id}."))
        else:
            display(Markdown(f"### No hero data available for {account}, Hero ID: {hero_id}."))
    else:
        display(Markdown("### Hero ID is unknown; cannot fetch hero data."))

# Function to get the top player across all classes and display their info
def get_top_player_info(season_id):
    leaderboard_data = fetch_leaderboard_data_for_all_classes(season_id)
    if not leaderboard_data.empty:
        # Sort by Rift Time in milliseconds (ascending)
        leaderboard_data_sorted = leaderboard_data.sort_values(by='Rift Time (ms)', ascending=True)

        # Iterate over the leaderboard to find a player with a valid HeroId
        for idx, top_player in leaderboard_data_sorted.iterrows():
            battletag = top_player['BattleTag']
            hero_id = top_player['HeroId']

            if battletag:
                battletag = battletag.replace("#", "-")
            else:
                battletag = "Unknown"

            if pd.notnull(hero_id):
                hero_id = int(hero_id)
                hero_class_slug = top_player['Class'].lower().replace(' ', '-')
                display(Markdown(f"### Top Player Across All Classes: {battletag}"))
                # Display the player's information
                display_player_info(battletag, hero_id, hero_class_slug)
                break  # Exit the loop after finding the first valid player
            else:
                continue  # Move to the next player
    else:
        display(Markdown("### No leaderboard data available"))

# Execute the function to display the top player's information
# Replace '23' with the desired season number
get_top_player_info(32)
```

## 9. Display Top Player's Hero Skills

In this section, we fetch and display the skills of the top player's hero to gain insights into their build and abilities.



```python
def display_hero_skills(account, hero_id):
    # Fetch hero data
    if hero_id != "Unknown":
        hero_data = fetch_hero_data(account, hero_id)
        if hero_data:
            # Display hero skills
            skills = hero_data.get('skills', {})
            active_skills = skills.get('active', [])
            passive_skills = skills.get('passive', [])

            # Display Active Skills
            if active_skills:
                active_skills_flattened = []
                for skill_slot in active_skills:
                    if 'skill' in skill_slot:
                        skill = skill_slot['skill']
                        rune = skill_slot.get('rune', {})
                        active_skills_flattened.append({
                            'Skill Name': skill.get('name', 'Unknown'),
                            'Rune Name': rune.get('name', ''),
                            'Description': skill.get('description', 'No description'),
                            'Rune Description': rune.get('description', '')
                        })
                display(Markdown(f"### Active Skills for {hero_data.get('name', 'Hero')}"))
                display(pd.DataFrame(active_skills_flattened))
            else:
                display(Markdown(f"No active skills found for {hero_data.get('name', 'Hero')}"))

            # Display Passive Skills
            if passive_skills:
                passive_skills_flattened = []
                for skill_slot in passive_skills:
                    if 'skill' in skill_slot:
                        skill = skill_slot['skill']
                        passive_skills_flattened.append({
                            'Skill Name': skill.get('name', 'Unknown'),
                            'Description': skill.get('description', 'No description')
                        })
                display(Markdown(f"### Passive Skills for {hero_data.get('name', 'Hero')}"))
                display(pd.DataFrame(passive_skills_flattened))
            else:
                display(Markdown(f"No passive skills found for {hero_data.get('name', 'Hero')}"))
        else:
            display(Markdown(f"### No hero data available for {account}, Hero ID: {hero_id}."))
    else:
        display(Markdown("### Hero ID is unknown; cannot fetch hero data."))


# Function to get the top player's skills
def get_top_player_skills(season_id):
    leaderboard_data = fetch_leaderboard_data_for_all_classes(season_id)

    if not leaderboard_data.empty:
        # Sort by Rift Time in milliseconds (ascending)
        leaderboard_data_sorted = leaderboard_data.sort_values(by='Rift Time (ms)', ascending=True)

        # Get the top player
        top_player = leaderboard_data_sorted.iloc[0]

        battletag = top_player['BattleTag']
        if battletag:
            battletag = battletag.replace("#", "-")
        else:
            battletag = "Unknown"

        hero_id = int(top_player['HeroId']) if top_player['HeroId'] else "Unknown"
        display(Markdown(f"### Top Player's Hero Skills: {battletag}"))

        # Display the hero's skills
        display_hero_skills(battletag, hero_id)
    else:
        display(Markdown("### No leaderboard data available"))

# Execute the function to display the top player's hero skills
get_top_player_skills(32)
```

## 10. Display First Monk Player Information

We will find the first Monk player from the leaderboard and display their information.


```python
# Function to convert Rift Time from milliseconds to mm:ss format
def convert_time_to_mmss(milliseconds):
    if milliseconds is not None:
        seconds = milliseconds / 1000  # Convert milliseconds to seconds
        minutes, seconds = divmod(seconds, 60)  # Get minutes and seconds
        return f"{int(minutes):02}:{int(seconds):02}"  # Format as mm:ss
    return "N/A"

# Function to fetch leaderboard data for Monk
def fetch_leaderboard_monk():
    url = f"{BASE_URL}/leaderboard?season_id=32&leaderboard_name=rift-monk"
    response = requests.get(url)
    if response.status_code == 200:
        return response.json()
    else:
        print(f"Failed to fetch leaderboard data: {response.status_code}")
        return []

def display_first_monk_player_info():
    leaderboard_data = fetch_leaderboard_monk()
    if leaderboard_data:
        # Normalize leaderboard JSON into a DataFrame
        df_leaderboard = pd.json_normalize(leaderboard_data['row'])
    
        # Get the first player
        first_player = df_leaderboard.iloc[0]
        first_player_data = first_player['data']
        player_data_list = first_player['player']
        
        # Convert 'data' list to a dictionary for easy access
        row_data = {item['id']: item for item in first_player_data}
        
        # Try to get the BattleTag from row_data
        battletag = row_data.get('BattleTag', {}).get('string', None)
        
        # If BattleTag is still None, attempt to get it from player data
        if not battletag and player_data_list:
            player_data = player_data_list[0]['data']
            player_info = {item['id']: item for item in player_data}
            battletag = player_info.get('HeroBattleTag', {}).get('string', None)
        
        if not battletag:
            battletag = 'Unknown'
        else:
            battletag = battletag.replace("#", "-")
        
        # Extract HeroId from player data
        if player_data_list:
            player_data = player_data_list[0]['data']
            player_info = {item['id']: item for item in player_data}
            hero_id = player_info.get('HeroId', {}).get('number', 'Unknown')
        else:
            hero_id = 'Unknown'
        
        # Extract Rift Time in milliseconds
        rift_time_ms = row_data.get('RiftTime', {}).get('timestamp', None)
        # Convert Rift Time to mm:ss format
        rift_time_formatted = convert_time_to_mmss(rift_time_ms)
        
        # Display player information
        display(Markdown(f"### First Monk Player: {battletag}"))
        display(Markdown(f"**Rift Time:** {rift_time_formatted}"))
        
        # Use the retrieved battletag and hero_id to fetch and display player info
        display_player_info(battletag, hero_id, 'monk')
    else:
        display(Markdown("### No leaderboard data available."))

# Example: Display player info based on the leaderboard selection
display_first_monk_player_info()
```

## 11. Fetch and Display Artisan Data

We will fetch artisan data and allow users to select different tiers to view recipes.


```python
# Function to fetch artisan data
def fetch_artisan(artisan_slug):
    url = f"{BASE_URL}/artisan/{artisan_slug}"
    return fetch_data(url)

# Function to display structured artisan data with tier selection
def display_artisan_data(artisan_slug, selected_tier=1):
    artisan_data = fetch_artisan(artisan_slug)
    if artisan_data:
        display(Markdown(f"## Artisan Data for {artisan_data['name']}"))

        # Extract and display general artisan info
        artisan_info = {
            "Name": artisan_data.get('name', 'N/A'),
            "Portrait": artisan_data.get('portrait', 'N/A'),
            "Slug": artisan_data.get('slug', 'N/A'),
        }
        display(pd.DataFrame([artisan_info]))

        # Fetch training data for the selected tier
        training_data = artisan_data.get('training', {}).get('tiers', [])
        
        if selected_tier > len(training_data) or selected_tier < 1:
            display(Markdown(f"### No data for Tier {selected_tier}. Available tiers: 1 to {len(training_data)}"))
            return
        
        # Get the specific tier's recipes
        tier_data = training_data[selected_tier - 1]
        tier_recipes = tier_data.get('taughtRecipes', [])
        
        if tier_recipes:
            display(Markdown(f"### Recipes for {artisan_data['name']} - Tier {selected_tier}"))
            training_info = []
            for recipe in tier_recipes:
                training_info.append({
                    "Recipe Name": recipe.get('name', 'N/A'),
                    "Cost": recipe.get('cost', 'N/A'),
                    "Produced Item": recipe.get('itemProduced', {}).get('name', 'N/A'),
                    "Reagents": ', '.join([r['item']['name'] for r in recipe.get('reagents', [])])
                })
            display(pd.DataFrame(training_info))
        else:
            display(Markdown(f"No recipes available for Tier {selected_tier}."))
    else:
        display(Markdown("### No artisan data available"))

# Dropdown widget for selecting artisan
artisan_dropdown = widgets.Dropdown(
    options=['blacksmith', 'jeweler'],
    value='blacksmith',
    description='Artisan:'
)

# Dropdown widget for selecting tier
tier_dropdown = widgets.IntSlider(
    value=1,
    min=1,
    max=12,
    step=1,
    description='Tier:'
)

# Function to update the display based on dropdown values
def update_artisan_data(artisan, tier):
    display_artisan_data(artisan, tier)

# Link the dropdowns to the update function
ui = widgets.VBox([artisan_dropdown, tier_dropdown])
out = widgets.interactive_output(update_artisan_data, {'artisan': artisan_dropdown, 'tier': tier_dropdown})

# Display the dropdowns and output
display(ui, out)
```

## 12. Fetch and Display Most Used Items

We will fetch and display the most used items for a selected class and season.


```python
# Fetch seasons data to populate the season dropdown
seasons_data = fetch_seasons()

# Extract season numbers from the 'href' field and clean the '?namespace' part
if 'season' in seasons_data:
    seasons = [season['href'].split('?')[0].split('/')[-1] for season in seasons_data['season']]  # Extract and clean season numbers
else:
    print("Unexpected data structure:", seasons_data)
    seasons = []  # Handle the case where no seasons are fetched

# Ensure there are available seasons
if seasons:
    season_dropdown = widgets.Dropdown(
        options=seasons,
        value=seasons[-1],  # Default to the last season
        description='Season:',
        disabled=False,
    )
else:
    print("No seasons available.")

# Function to fetch most used items from the API based on season and class
def fetch_most_used_items(season_id, class_slug):
    url = f"{BASE_URL}/most_used_items?season_id={season_id}&class_slug={class_slug}"
    response = requests.get(url)

    if response.status_code == 200:
        return response.json()  # Return the items data
    else:
        print(f"Failed to fetch most used items: {response.status_code}")
        return {}

# Function to display the most used items in a DataFrame format
def display_most_used_items(items_data):
    if not items_data:
        display(Markdown("No items data found for the selected class and season."))
        return

    item_rows = []
    for slot, item_info in items_data.items():
        item = item_info['item']
        count = item_info['count']

        # Check if item is a dictionary
        if not isinstance(item, dict):
            print(f"Item is not a dictionary for slot {slot}, skipping.")
            continue

        # Extract item name and other details
        item_name = item.get('name', 'Unknown')
        armor = item.get('armor', 'N/A')
        augmentation = item.get('augmentation', 'N/A')
        primary_attrs = ', '.join(item.get('attributes', {}).get('primary', [])) if 'attributes' in item else "N/A"
        secondary_attrs = ', '.join(item.get('attributes', {}).get('secondary', [])) if 'attributes' in item else "N/A"

        # Process gems if available
        gems_list = []
        if 'gems' in item:
            for gem in item['gems']:
                gem_name = gem['item']['name']
                gem_attrs = ', '.join(gem.get('attributes', []))
                gems_list.append(f"{gem_name}: {gem_attrs}")
            gems = '; '.join(gems_list)
        else:
            gems = "N/A"

        flavor_text = item.get('flavorText', 'N/A')

        # Add item details to rows
        item_rows.append({
            "Slot": slot.capitalize(),
            "Item Name": item_name,
            "Usage Count": count,
            "Armor": armor,
            "Augmentation": augmentation,
            "Primary Attributes": primary_attrs,
            "Secondary Attributes": secondary_attrs,
            "Gems": gems,
            "Flavor Text": flavor_text
        })

    # Create a DataFrame to display the items
    df_items = pd.DataFrame(item_rows)
    display(df_items)

    # Plot the usage counts
    plt.figure(figsize=(12, 6))
    sns.barplot(
        x='Slot', 
        y='Usage Count', 
        data=df_items.sort_values('Usage Count', ascending=False), 
        hue='Slot', 
        palette='viridis', 
        dodge=False, 
        legend=False
    )
    plt.title('Most Used Items by Slot')
    plt.xlabel('Item Slot')
    plt.ylabel('Usage Count')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()




# Function to update and display the most used items
def update_most_used_items(season_id, class_slug):
    # Fetch and display most used items
    most_used_items_data = fetch_most_used_items(season_id, class_slug)
    display(Markdown(f"### Most Used Items for {class_slug.replace('rift-', '').capitalize()} in Season {season_id}"))
    display_most_used_items(most_used_items_data)

# Interactive widget for class selection
classes = ['rift-wizard', 'rift-barbarian', 'rift-dh', 'rift-monk', 'rift-necromancer', 'rift-wd', 'rift-crusader']

# Dropdown for class selection
class_dropdown = widgets.Dropdown(
    options=classes,
    value=classes[0],  # Default to the first class
    description='Class:',
    disabled=False,
)

# Update the interact function to call the new item display function
interact(update_most_used_items, season_id=season_dropdown, class_slug=class_dropdown)
```
